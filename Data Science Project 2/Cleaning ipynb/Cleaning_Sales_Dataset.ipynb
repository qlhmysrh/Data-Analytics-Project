# **Check outliers**

```python
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named 'df'
# Replace 'df' with your actual DataFrame name

# Select the columns you want to plot
columns_to_plot = ['Price', 'Quantity', 'Tax', 'TotalPrice', 'GIncome']

# Create a boxplot
plt.figure(figsize=(10, 6))
sales.boxplot(column=columns_to_plot)
plt.title('Boxplot of Price, Quantity, Tax, TotalPrice, and GIncome')
plt.ylabel('Value')
plt.xlabel('Variable')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```
